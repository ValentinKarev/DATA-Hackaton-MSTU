```python
import numpy as np
import matplotlib.pyplot as plt

def cluster_points_with_balance(points, max_x_diff=3, target_clusters=60):
    # Вычисляем общую сумму y и целевую сумму для каждого кластера
    total_y = np.sum(points[:, 1])
    target_y_per_cluster = total_y / target_clusters

    clusters = []  # Список кластеров

    for point in tqdm(points):
        x, y = point
        added_to_cluster = False

        # Проверяем, можно ли добавить точку в один из существующих кластеров
        best_fit_index = -1
        best_fit_difference = float('inf')

        for i, cluster in enumerate(clusters):
            cluster_x_min = np.min(cluster[:, 0])  # Минимальное значение x в кластере
            cluster_y_sum = np.sum(cluster[:, 1])  # Сумма y в кластере

            # Проверяем условия: разница по x и новая сумма y не превышает целевую
            if (x - cluster_x_min <= max_x_diff) and (cluster_y_sum + y <= target_y_per_cluster * 1.2):
                # Вычисляем разницу с целевой суммой
                difference = abs((cluster_y_sum + y) - target_y_per_cluster)

                # Ищем кластер, который лучше всего подходит
                if difference < best_fit_difference:
                    best_fit_index = i
                    best_fit_difference = difference

        # Если нашли подходящий кластер, добавляем точку
        if best_fit_index != -1:
            clusters[best_fit_index] = np.vstack([clusters[best_fit_index], point])
            added_to_cluster = True

        # Если точку нельзя добавить ни в один из существующих кластеров
        if not added_to_cluster:
            clusters.append(np.array([point]))  # Создаем новый кластер

    return clusters


# Генерация случайных данных для примера
np.random.seed(42)  # Для воспроизводимости

# Кластеризация точек
clusters = cluster_points_with_balance(a)

# Вывод результатов
print(f"Количество кластеров: {len(clusters)}")
for i, cluster in enumerate(clusters):
    print(f"Кластер {i+1}: {len(cluster)} точек, сумма y = {np.sum(cluster[:, 1]):.2f}")
```
